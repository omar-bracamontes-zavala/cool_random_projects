# Black Jack

```python
# Set up the deck
suits = ('Hearts','Diamonds','Spades','Clubs')
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}

playing = True
```

```python
# Set up Card class
class Card:
    '''
    ABLE TO UNDERSTAN SUIT, RANK, VALUE
    '''
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + ' of ' + self.suit
```

```python
# Set up Deck class
from random import shuffle

class Deck:
    '''
    ABLE TO:
    CREATE 52 CARDS AND HOLD THEM AS A LIST, 
    SHUFFLE A DECK,
    DEAL CARDS FROM THE DECK OBJECT
    '''
    
    def __init__(self):
        self.deck = []
        
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
                
    def __str__(self):
        deck_comp = ''  
        for card in self.deck:
            deck_comp += '\n '+card.__str__() 
        return 'The deck has:' + deck_comp
    
    def shuffle(self):
        shuffle(self.deck)
        
    def deal(self):
        if len(self.deck) > 0:
            return self.deck.pop()
        else:
            print('No more cards in deck!')
```

```python
# Set up Hand class
class Hand:
    '''
    HOLD CARDS DEALT FROM THE DECK AND CALCULATE THE VALUE 
    OF THOSE CARDS. IT ALSO ADJUST ACES VALUE
    '''
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
    
    def add_card(self,card):
        self.cards.append(card)
        self.value += card.value
        
        # track aces
        if card.rank == 'Ace':
            self.aces += 1
    
    def adjust_for_ace(self):
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
# Set up Chips class
class Chips:
    '''
    TRACK PLAYER'S STARTING CHIPS, 
    BETS AND ONGOINGWINNINGS
    '''
    def __init__(self):
        self.total = 100000000000  # This can be set to a default value or supplied by a user input
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
        print(f'You win the bet! ${self.bet} has been added\nTotal: ${self.total}\n')
    
    def lose_bet(self):
        self.total -= self.bet
        print(f'You lose the bet! ${self.bet} has been lost\nTotal: ${self.total}\n')
```

```python
# Can i bet?
def take_bet():
    '''
    IF THE PLAYES HAS ENOUGH FOUNDS,
    TAKE THE BET
    '''
    while True:
        try:
            player_chips.bet = int(input('How many chips would you like to bet? $'))
        except:
            print('Whoops, thats not a number. Please provide an integer \n')
        else:
            if player_chips.bet > player_chips.total:
                print(f'Sorry! You dont have enough chips. You have ${player_chips.total}')
            else:
                break
```

```python
# Hit
def hit(deck,hand):
    '''
    EITHER PLAYER CAN TAKE  HITS UNTIL THEY BUST
    '''
    hand.add_card(deck.deal())
    hand.adjust_for_ace()
```

```python
def hit_or_stand(deck,hand):
    '''
    ASK TO THE PLAYER IF HE WANTS TO HIT OR STAND
    '''
    global playing  # to control an upcoming while loop
    
    while True:
        player_choice = input('Do you want to hit or stand? ')

        if player_choice[0].lower() == 'h':
            hit(deck,hand)
            
        elif player_choice[0].lower() == 's':
            print('Player stands. Dealers turn')
            playing = False
            
        else:
            print('Sorry, I dont understand that. Please enter h or s!\n')
            continue
            
        break
```

```python
def show_some(player,dealer):
    print("\nDealer's Hand:")
    print(" <card hidden>")
    print('',dealer.cards[1])  
    print("\nPlayer's Hand:", *player.cards, sep='\n ')
    print('_'*20)
    
def show_all(player,dealer):
    print("\nDealer's Hand:", *dealer.cards, sep='\n ')
    print("Dealer's Hand =",dealer.value)
    print("\nPlayer's Hand:", *player.cards, sep='\n ')
    print("Player's Hand =",player.value)
    print('_'*20)
```

```python
def player_busts(player,dealer,chips):
    print('\nBUST PLAYER!')
    chips.lose_bet()
    
def player_wins(player,dealer,chips):
    print('\nPLAYER WINS!')
    chips.win_bet()

def dealer_busts(player,dealer,chips):
    print('\nPLAYER WINS! DEALER BUSTED!')
    chips.win_bet()
    
def dealer_wins(player,dealer,chips):
    print('\nDEALER WINS!')
    chips.lose_bet()
    
def push(player,dealer):
    print('\nDealer and player tie! PUSH!')
```
